```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection

X, y = datasets.load_linnerud(return_X_y=True)
print(X.shape)
print("Feature")
print(X[0])
print("Target: ")
print(y[0])
```

```python
X = X[:, 2] # take only one feature
X = X.reshape((-1,1)) # reshape to 2D array (x rows, 1 column)
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
print(len(X_test))
```

```python
# train the model
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test[3], y_test[3], color='black')
plt.scatter(X_test[3], y_pred[3], color='red', linewidth=5)

plt.xlabel('Physiological Weight, Waist and Pulse.')
plt.ylabel('Exercise Chins, Situps and Jumps.')
plt.title('A Graph Plot Showing 3 physiological variables Against 3 exercise')
plt.show()
```
